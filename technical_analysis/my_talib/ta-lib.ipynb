```python
import talib
import numpy as np
```

```python
c = numpy.random.randn(100)
```

```python
k, d = talib.STOCHRSI(c)
```

```python
talib.SMA(c)
```

```python
talib.RSI(c, 5)
```

```python
def get_rand():
    return 5 * np.random.random_sample(100) + 1
```

```python
c_open = get_rand()
c_high = get_rand()
c_low = c_high - get_rand()
c_low[c_low < 0] = 0
c_close = get_rand()
```

```python
talib.MACD(c_close)
```

```python
talib.CDLHAMMER(c_open, c_high, c_low, c_close)
```

```python
import plotly.graph_objects as go
import pandas as pd
```

```python
df = pd.DataFrame()
df["Open"] = c_open
df["High"] = c_high
df["Low"] = c_low
df["Close"] = c_close
df
```

```python
fig = go.Figure(data=go.Ohlc(x=df.index,
                             open=df['Open'],
                             high=df['High'],
                             low=df['Low'],
                             close=df['Close']))
  
# show the figure
fig.show()
```
